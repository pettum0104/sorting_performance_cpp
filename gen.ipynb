```python
prepayment_ranges = (0.1, 0.6)

for num_samples in range(100, 100001, 8000):
    data = []
    
    for _ in range(num_samples):
        category = random.choice(categories)
        if category in base_services:
            base_service = random.choice(base_services[category])
            service_name_suffix = random.choice(service_name_suffixes)
            service_name = base_service + service_name_suffix
    
            if category in price_ranges and category in duration_ranges:
                price = round(random.uniform(*price_ranges[category]), 2)
                duration = random.randint(*duration_ranges[category])
                prepayment = round(price * random.uniform(*prepayment_ranges), 2)
                data.append([service_name, price, duration, prepayment])
            else:
                 continue
        else:
            continue
    
    
    df = pd.DataFrame(data, columns=["Название услуги", "Ориентировочная стоимость", "Срок исполнения (дни)", "Размер предоплаты"])
    csv_file_path = f"datasets/it_services_dataset_diverse_{num_samples}.csv"
    df.to_csv(csv_file_path, index=False, encoding='utf-8')
```
