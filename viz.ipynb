```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np

df = pd.read_csv('results/timing_results_bvg_all.csv')

sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))
line_plot = sns.lineplot(
    data=df,
    x='DatasetSize',
    y='TimeMilliseconds',
    hue='Algorithm',
    marker='o'
)

plt.title('Сравнение времени выполнения алгоритмов сортировки', fontsize=16)
plt.xlabel('Размер набора данных (количество записей)', fontsize=12)
plt.ylabel('Временя выполнения (мс)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Алгоритм', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout(rect=[0, 0, 0.85, 1])

plt.show()
```

```python
df.TimeMilliseconds = np.log(df.TimeMilliseconds + 1)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))
line_plot = sns.lineplot(
    data=df,
    x='DatasetSize',
    y='TimeMilliseconds',
    hue='Algorithm',
    marker='o'
)

plt.title('Сравнение времени выполнения алгоритмов сортировки', fontsize=16)
plt.xlabel('Размер набора данных (количество записей)', fontsize=12)
plt.ylabel('Логарифм временени выполнения', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Алгоритм', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout(rect=[0, 0, 0.85, 1])

plt.show()
```

**Вывод:**

По графикам видно, что сортировки пузырьком, вставками и шейкер-сортировка ведут себя очень похоже и их время выполнения быстро растет с увеличением данных. Это ожидаемо для алгоритмов со сложностью $N^2$. Из этой тройки, шейкер-сортировка кажется чуть быстрее на больших данных, вставками - на небольших, но в целом, они все очень медленные для больших наборов данных. Их стоит использовать разве что для учебы или если данных совсем мало.

Время выполнения *std::sort* растет значительно медленнее, она использует гораздо более эффективный алгоритм со сложностью $NlogN$, который еще и сильно оптимизирован.

**Итог**: Если вам нужно сортировать что-то для реальной задачи, особенно с заметным количеством данных, всегда используйте std::sort. Пузырьком, вставками и шейкер-сортировка – это, по сути, только для понимания принципов работы и для самых маленьких списков.
